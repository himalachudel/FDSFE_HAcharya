**Now moving ahead with KS test**

```python
stats.ks_2samp(happiness["Generosity"],happiness["Generosity"])
```

**The P value of 1.0 is way higher than 0.05 at 95% CI. Thus null hypothesis is thus not rejected. The reason for the highest value possible of p is  because we are trying to compare two same distribution of values.**

```python
stats.ks_2samp(happiness["Happiness Score"],happiness["Economy (GDP per Capita)"])
```

**The P value of 3.34X10^-94 is way smaller than 0.05 at 95% CI. Thus null hypothesis is rejected.**
